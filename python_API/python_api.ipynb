```python
from config import API_KEY
import requests
import json
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import random
from citipy import citipy
```

```python
query_url = f'https://api.openweathermap.org/data/2.5/weather?appid={API_KEY}'
units = "imperial"
```

```python
lat = random.uniform(-90, 90)
lng = random.uniform(-180, 180)
city = citipy.nearest_city(lat,lng)
max_temp = requests.get(query_url + "&q=" + city.city_name + "&units=" + units).json()["main"]["temp_max"]
humidity = requests.get(query_url + "&q=" + city.city_name + "&units=" + units).json()["main"]["humidity"]
cloudiness = requests.get(query_url + "&q=" + city.city_name + "&units=" + units).json()["clouds"]["all"]
wind_speed = requests.get(query_url + "&q=" + city.city_name + "&units=" + units).json()["wind"]["speed"]
url2 = requests.get(query_url + "&q=" + city.city_name + "&units=" + units).url

print(lat)
print(lng)
print(city)
print(max_temp)
print(humidity)
print(cloudiness)
print(wind_speed)
print(url2)
```

```python
# set up lists to hold reponse info
ran_cities= []

i=1
while i<= 500:
    try:
        flag = True
        lat = random.uniform(-90, 90)
        lng = random.uniform(-180, 180)
        city = citipy.nearest_city(lat,lng)
        max_temp = requests.get(query_url + "&q=" + city.city_name + "&units=" + units).json()["main"]["temp_max"]
        humidity = requests.get(query_url + "&q=" + city.city_name + "&units=" + units).json()["main"]["humidity"]
        cloudiness = requests.get(query_url + "&q=" + city.city_name + "&units=" + units).json()["clouds"]["all"]
        wind_speed = requests.get(query_url + "&q=" + city.city_name + "&units=" + units).json()["wind"]["speed"]
        url2 = requests.get(query_url + "&q=" + city.city_name + "&units=" + units).url
        ran_cities.append((lat,lng,city.city_name,city.country_code, max_temp, humidity, cloudiness, wind_speed, url2))
        print("Processed: " + str(i) + " " + str([lat,lng,city.city_name,city.country_code, max_temp, humidity, cloudiness, wind_speed, url2]))
        
    except KeyError:
        pass
        flag = False
    if flag == True:
        i = i + 1

# rand dataframe
weather_df = pd.DataFrame(ran_cities,columns=["latitude","longitude","city","country","max_temp","humidity","cloudiness","wind_speed","url"])
weather_df.head()
```

```python
weather_df = pd.DataFrame(ran_cities,columns=["latitude","longitude","city","country","max_temp","humidity","cloudiness","wind_speed","url"])
weather_df.head()
```
