# Scrape: Mars Hemispheres

```python
# (1/5) Create: Browser object
browser = Browser("chrome", executable_path="chromedriver.exe", headless=False)

# (2/5) Create: URL obj
url_usgs = "https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars"

# (3/5) Visit: Website
browser.visit(url_usgs)

# Convert: Current page to html
html = browser.html

# Create: Beautiful Soup object
soup_hemi = bs(html, "html.parser")
```

### ========== Testing codes: Start ==========

```python
print(soup_hemi.find_all("div", class_="item")[0].prettify())
```

```python
# Scrape: URL paths
for i in soup_hemi.find_all("div", class_="item"):
    print(i.find("a", class_="itemLink product-item").get("href"))
```

```python
# Scrape: The name
for i in soup_hemi.find_all("div", class_="item"):
    print(i.find("h3").text)
```

### ========== Testing codes: End ==========

```python
list_hemis = []

# Scrape: The hemisphere names
for i in soup_hemi.find_all("div", class_="item"):
    
    # Create: Dict
    dict_hemis = {}
    
        # Finding the hemisphere name 
    
    # Add: Hemisphere names to list
    dict_hemis["hemi_name"] = i.find("h3").text
   
        # Finding the hemisphere .jpg
    
    # Find: The partial URL path
    url_hemi_loop = i.find("a", class_="itemLink product-item").get("href")
    
    # Visit: Website
    browser.visit(f'https://astrogeology.usgs.gov/{url_hemi_loop}')
       
    # Click: By id
    browser.find_by_id("wide-image-toggle").click()
    
    # Convert: Current page to html
    html = browser.html
    
    # Create: Beautiful Soup object
    soup_hemi_loop = bs(html, "html.parser")
    
    # Scrape: JPG
    jpg_loop = soup_hemi_loop.find("img", class_="wide-image").get("src")
    
    # Add: JPG to list
    dict_hemis["jpg_url"] = jpg_loop
    
    list_hemis.append(dict_hemis)
    
list_hemis
```
